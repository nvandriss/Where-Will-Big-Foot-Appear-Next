```python
import pandas as pd
import numpy as np
```

```python
midwestpath='Resources/MidwestCamp.csv'
northeastpath='Resources/NortheastCamp.csv'
southpath='Resources/SouthCamp.csv'
southwestpath='Resources/SouthwestCamp.csv'
westpath='Resources/WestCamp.csv'
```

```python
midwest_df=pd.read_csv(midwestpath, engine='python')
northeast_df=pd.read_csv(northeastpath, engine='python')
south_df=pd.read_csv(southpath, engine='python')
southwest_df=pd.read_csv(southwestpath, engine='python')
west_df=pd.read_csv(westpath, engine='python')
```

```python
campgrounds_df = midwest_df.append(northeast_df)
campgrounds_df = campgrounds_df.append(south_df)
campgrounds_df = campgrounds_df.append(southwest_df)
campgrounds_df = campgrounds_df.append(west_df)
campgrounds_df
```

```python
campgrounds_df.dropna()

campgrounds_df = campgrounds_df[['Lat', 
                            'Long', 
                            'Code', 
                            'Campground', 
                            'Type', 
                            'City', 
                            'State']]
campgrounds_df
```

```python
campgrounds = campgrounds_df.sort_values('State')
campgrounds.head()
```

```python
#find the total number of campgrounds
campgrounds_rank = campgrounds.groupby("State").count()
campgrounds_rank
```

```python
campgrounds_ranked1 = campgrounds_rank.sort_values(by="Campground", ascending=False)

campgrounds_ranked_df = campgrounds_ranked1.reset_index()

campgrounds_ranked_df
```
