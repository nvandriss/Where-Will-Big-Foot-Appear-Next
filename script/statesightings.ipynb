```python
import pandas as pd 
import os
```

```python
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
long_langdata=pd.read_csv(os.path.join("Resources", "bfro_locations.csv"))
```

```python
long_langdata
```

```python
long_langdata['geom'] = long_langdata['latitude'].map(str) + ',' + long_langdata['longitude'].map(str)
long_langdata['geom']
```

```python
geoLoc = Nominatim(user_agent='GetLoc', timeout=10)
rgeocode = RateLimiter(geoLoc.reverse, min_delay_seconds=0.001)
```

```python
states=[]
countries=[]
for loc in long_langdata['geom']:
    try:
        gl=geoLoc.reverse(loc)
        address=gl.raw['address']
        state = address.get('state', '')
        country = address.get('country', '')
        print(state,country)
        states.append(state)
        countries.append(country)
    except:
        states.append('')
        countries.append('')
        
```

```python
long_langdata.head()
```

```python
long_langdata['timestamp'] = pd.to_datetime(long_langdata['timestamp'],infer_datetime_format=True)
long_langdata['year'] = pd.DatetimeIndex(long_langdata['timestamp']).year
long_langdata.head()
```

```python
sightings_df=pd.DataFrame({'title':long_langdata['title'],
'classification':long_langdata['classification'],
'year': long_langdata['year'],
'latitude':long_langdata['latitude'],
'longitude':long_langdata['longitude'],
'State': states,
'country':countries
})
sightings_df
```

```python
sightings_df=sightings_df.drop(sightings_df.loc[sightings_df['State']== ''].index)
```

```python
sightings_df=sightings_df.drop(sightings_df.loc[sightings_df['country']== ''].index)
sightings_df
```

```python
sightings_df=sightings_df.drop(sightings_df.loc[sightings_df['country']== 'Canada'].index)
sightings_df
```
