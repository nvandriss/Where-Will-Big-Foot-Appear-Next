```python
import pandas as pd 
import os
```

```python
rawdata = pd.read_csv(os.path.join("Resources", "bfro_reports_geocoded.csv"))
long_langdata=pd.read_csv(os.path.join("Resources", "bfro_locations.csv"))
```

```python
long_langdata
```

```python
long_langdata['timestamp'].max()
long_langdata['timestamp'] = pd.to_datetime(long_langdata['timestamp'],infer_datetime_format=True)
long_langdata['year'] = pd.DatetimeIndex(long_langdata['timestamp']).year
long_langdata
```

```python
rawdata=rawdata.dropna()
rawdata.columns
```

```python
sightings=rawdata[['state','latitude', 'latitude','classification','title','date']]
sightings
```

```python
countsite=sightings.groupby('state',0).count()
countsite=countsite.reset_index()
countperstate=countsite[['state','title']]
count0={'state':'Hawaii',
'title':0}
countperstate=countperstate.append(count0,ignore_index = True)
count1={'state':'Nevada',
'title':0}
countperstate=countperstate.append(count1,ignore_index = True)
count2={'state':'North Dakota',
'title':0}
countperstate=countperstate.append(count2,ignore_index = True)
count3={'state':'Wyoming',
'title':0}
countperstate=countperstate.append(count3,ignore_index = True)
count4={'state':'Alabama',
'title':0}
countperstate=countperstate.append(count4,ignore_index = True)
count5={'state':'DC',
'title':0}
countperstate=countperstate.append(count5,ignore_index = True)
countperstate

```

```python
countperstate
```

```python
countperstate.to_csv('countperstate.csv')
sightings.to_csv('sightings.csv')
long_langdata.to_csv('long_langdata.csv')
```
